```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/OMF_data_final/uploaded_data_2/Dataset_2_csv.csv')
df = df[df.columns[1:]]
df = df.iloc[::-1]
```

```python
window = 30
num_days = len(df.index)
num_companies = df.columns.size
total_rows = num_days - window + 1
```

```python
prices = df.to_numpy()
```

```python
return_rate = (prices[1:] - prices[:-1])/prices[:-1]
```

```python
w_0_21_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_21 beta')
w_0_22_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_22 beta')
w_0_23_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_23 beta')
w_0_24_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_24 beta')
w_0_25_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_25 beta')
w_0_26_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_26 beta')
w_0_27_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_27 beta')
# w_0_8_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with 0_8 beta')
# w_og_df = pd.read_csv('/content/drive/MyDrive/Weights without transaction cost/Weights with normal average')
```

```python
w_0_21_df = w_0_21_df[:-1]
w_0_22_df = w_0_22_df[:-1]
w_0_23_df = w_0_23_df[:-1]
w_0_24_df = w_0_24_df[:-1]
w_0_25_df = w_0_25_df[:-1]
w_0_26_df = w_0_26_df[:-1]
w_0_27_df = w_0_27_df[:-1]
# w_0_8_df = w_0_8_df[:-1]
# w_og_df = w_og_df[:-1]
```

```python
# w_og_df = w_og_df.iloc[:,1:]
w_0_21_df = w_0_21_df.iloc[:,1:]
w_0_22_df = w_0_22_df.iloc[:,1:]
w_0_23_df = w_0_23_df.iloc[:,1:]
w_0_24_df = w_0_24_df.iloc[:,1:]
w_0_25_df = w_0_25_df.iloc[:,1:]
w_0_26_df = w_0_26_df.iloc[:,1:]
w_0_27_df = w_0_27_df.iloc[:,1:]
# w_0_8_df = w_0_8_df.iloc[:,1:]
```

```python
w_0_21 = w_0_21_df.to_numpy()
w_0_22 = w_0_22_df.to_numpy()
w_0_23 = w_0_23_df.to_numpy()
w_0_24 = w_0_24_df.to_numpy()
w_0_25 = w_0_25_df.to_numpy()
w_0_26 = w_0_26_df.to_numpy()
w_0_27 = w_0_27_df.to_numpy()
# w_0_8 = w_0_8_df.to_numpy()
# w_og = w_og_df.to_numpy()
```

```python
# return_og = []
return_0_21 = []
return_0_22 = []
return_0_23 = []
return_0_24 = []
return_0_25 = []
return_0_26 = []
return_0_27 = []
# return_0_8 = []
```

```python
day = []
```

```python
for i in range(window, total_rows):
  # r_og = w_og[i-window]@return_rate[i]
  r_0_21 = w_0_21[i-window]@return_rate[i]
  r_0_22 = w_0_22[i-window]@return_rate[i]
  r_0_23 = w_0_23[i-window]@return_rate[i]
  r_0_24 = w_0_24[i-window]@return_rate[i]
  r_0_25 = w_0_25[i-window]@return_rate[i]
  r_0_26 = w_0_26[i-window]@return_rate[i]
  r_0_27 = w_0_27[i-window]@return_rate[i]
  # r_0_8 = w_0_8[i-window]@return_rate[i]

  # return_og.append(r_og)
  return_0_21.append(r_0_21)
  return_0_22.append(r_0_22)
  return_0_23.append(r_0_23)
  return_0_24.append(r_0_24)
  return_0_25.append(r_0_25)
  return_0_26.append(r_0_26)
  return_0_27.append(r_0_27)
  # return_0_8.append(r_0_8)

  day.append(i-window+1)
```

```python
# plt.plot(day,return_og,'o',label='Original')
plt.plot(day,return_0_21, 'o', label='0.21')
plt.plot(day,return_0_22, 'o', label='0.22')
plt.plot(day,return_0_23, 'o', label='0.23')
plt.plot(day,return_0_24, 'o', label='0.24')
plt.plot(day,return_0_25, 'o', label='0.25')
plt.plot(day,return_0_26, 'o', label='0.26')
plt.plot(day,return_0_27, 'o', label='0.27')
# plt.plot(day,return_0_8, 'o', label='0.8')
plt.legend()
plt.show()
```
